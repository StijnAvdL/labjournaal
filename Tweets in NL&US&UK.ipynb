## Aantal Tweets per land per dag

```python
import pymongo
import matplotlib.pyplot as plt
import pandas as pd
import time

# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['app24236103']

songName = "take me to church"
today= time.strftime("%d/%m/%Y")

OneSong = db.testTable2.aggregate([
    {
        "$match": {
            "title": songName,
            "$or": [{"country_id": 840}, {"country_id": 528}, {"country_id": 826}]
        }
    },
    { "$unwind": "$plays"},
    {
        "$project": {
            "country_id": "$country_id",
            "plays": "$plays",
            "years": {"$year": "$plays.datum"}
        }
    },
    {
        "$match": {
            "years": {"$gt": 2014}
        }
    },
    {
        "$group": {
            "_id": {"day": {"$dayOfMonth": "$plays.datum"}, "month": {"$month": "$plays.datum"}, "year": {"$year": "$plays.datum"} },
            "plays": {"$push":
                {
                    "country_id": "$country_id", "count": "$plays.count"
                }
            }
        }
    },
    { "$sort": 
        {"_id.year": 1, "_id.month": 1, "_id.day": 1}
    },
    { "$project": {
           "_id": 0,
           "day": { "$substr": [ "$_id.day",   0, 2 ] }, 
           "month":  { "$substr": [ "$_id.month", 0, 2 ] }, 
           "year": { "$substr": [ "$_id.year",  0, 4 ] },
           "plays": "$plays"
        }     
    },
    { "$project": {  
           "date": {"$concat": ["$day", "-", "$month", "-", "$year"]},
           "plays": "$plays"
        }
    }
])

source = list(OneSong['result'])
print('recived data')
```

```python
countTweets=[[],[],[]]
datum=[]

for p, play in enumerate(source): 
    datum.append(play['date'])
    nl = 'false'
    us = 'false'
    uk = 'false'
    for c, count in enumerate(play['plays']):
        if count['country_id'] == 826:
            uk = 'true'
            countTweets[0].append(count['count'])
        elif  count['country_id'] == 840:
            us = 'true'
            countTweets[1].append(count['count'])
        elif   count['country_id'] == 528:
            nl = 'true'
            countTweets[2].append(count['count'])
    if uk == 'false':       
        countTweets[0].append(0)
    elif us == 'false':
        countTweets[1].append(0)
    elif  nl == 'false':
        countTweets[2].append(0)

print('convert data for plot... done')
```

```python
x=linspace(0,len(datum),len(datum))

nlLine = plt.plot(x, countTweets[2], 'r-', label='NL')
usLine = plt.plot(x, countTweets[1], 'b-', label='US')
ukLine = plt.plot(x, countTweets[0], 'g-', label='UK')
plt.ylabel('tweets song')
plt.locator_params(nbins=10)
plt.xticks(x,datum,rotation='vertical')
plt.xlabel('datum')
plt.title(songName + ': created at ' + today)
plt.legend()
plt.show()
```

## Aantal tweets per land per week

```python
songWeek = db.testTable2.aggregate([
    {
        "$match": {
            "title": songName,
            "$or": [{"country_id": 840}, {"country_id": 528}, {"country_id": 826}]
        }
    },
    { "$unwind": "$plays"},
    {
        "$project": {
            "country_id": "$country_id",
            "plays": "$plays",
            "years": {"$year": "$plays.datum"}
        }
    },
    {
        "$match": {
            "years": {"$gt": 2014}
        }
    },
    {
        "$group": {
            "_id": {"week": {"$week": "$plays.datum"}, "country_id": "$country_id"},
            "count": {"$sum": "$plays.count"}
 
        }
    },
    {
        "$group": {
            "_id": {"week": "$_id.week"},
            "plays": {"$push":
                {
                    "country_id": "$_id.country_id", "count": "$count"
                }
            }
 
        }
    },
    { "$sort": 
        {"_id.week": 1}
    },
    {
        "$project": {
            "_id": 0,
            "week": "$_id.week",
            "plays": "$plays"
        }
    }
])

source2 = list(songWeek['result'])
print('recived data')
```

```python
countTweets2=[[],[],[]]
datum2=[]

for p, play in enumerate(source2): 
    weekNumber = play['week'] + 1
    datum2.append(weekNumber)
    nl = 'false'
    us = 'false'
    uk = 'false'
    for c, count in enumerate(play['plays']):
        if count['country_id'] == 826:
            uk = 'true'
            countTweets2[0].append(count['count'])
        elif  count['country_id'] == 840:
            us = 'true'
            countTweets2[1].append(count['count'])
        elif   count['country_id'] == 528:
            nl = 'true'
            countTweets2[2].append(count['count'])
    if uk == 'false':       
        countTweets2[0].append(0)
    elif us == 'false':
        countTweets2[1].append(0)
    elif  nl == 'false':
        countTweets2[2].append(0)

print('convert data for plot... done')
```

```python
x=linspace(0,len(datum2),len(datum2))

nlLine = plt.plot(x, countTweets2[2], 'r-', label='NL')
usLine = plt.plot(x, countTweets2[1], 'b-', label='US')
ukLine = plt.plot(x, countTweets2[0], 'g-', label='UK')
plt.ylabel('tweets song each week')
plt.locator_params(nbins=10)
plt.xticks(x,datum2,rotation='vertical')
plt.xlabel('week number 2015')
plt.title(songName + ': created at ' + today)
plt.legend()
plt.show()
```
