## Wat is Pearson-r?

http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.pearsonr.html 

Pearson-r is een correlatie functie dat een score tussen de 1 en -1 terug geeft op basis van 2 cijfer reeksen. Waarbij 1 betekend dat de 2 cijfer reeksen perfect positieve correlatie hebben en bij -1 totaal niet. Daarnaast geeft het een p-value terug. Wat inhoud hoe betrouwbaar de correlatie score is. Waarbij hoe dichter bij de 0, des te vertrouwbaar de correlatie score is.

```python
a = [5, 7, 9, 5, 3, 2, 1]
b = [10, 14, 18, 10, 6, 4, 2]
st.pearsonr(a, b)
```

Reeks b loopt op dezelfde moment omhoog en omlaag met een meervoud van 2. Hierdoor wordt onderstelt dat reeksen exact hetzelfde zijn en de p-value geeft 0 aan waarmee bedoelt wordt dat je deze uitslag volledig kan vertrouwen. Hierdoor kan ik concluderen dat er geen normalisatie over de data uitgevoerd moet worden. 

```python
a = [5, 7, 9, 5, 3, 2, 1]
b = [10, 9, 5, 3, 2, 1, 0]
st.pearsonr(a, b)
```

Reeks a loopt eerst omhoog en vervolgen omlaag. Bij reeks b loopt de reeks alleen af. Hierdoor zie je dat de score een stuk lager is en je de uitslag wat minder kan vertrouwen. 

## NL & US vergelijken met pearson-r

```python
days = [[104,111],[126,132],[137,139]] # DE UITIENDELIJK DAGEN DIE GEBRUIKT WORDEN!!!!!!

# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['mozart']

print('connected to database')
```

```python
query = db.npCountryTopCC.find({'cc': 'NL'})
nlData = list(query)

query = db.npCountryTopCC.find({'cc': 'US'})
usData = list(query)
```

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

```python
def pearsonrCountry(countryOne, countryTwo, pValue, correctie, figure, pCount = False):
    days = [[104,111],[126,132],[137,139]]
    i = 0
    pearsonR = []
    
    pvaluecount = 0
    # land 1 doorlopen
    for songsOne in countryOne:
        # controleer of land 2 zong bezit
        for songsTwo in countryTwo:
            if songsOne['mbId'] == songsTwo['mbId']:
                i += 1
                # cross correlate muzieknummer van beide landen
                ccOne = getCountryPlays(songsOne, days)
                ccTwo = getCountryPlays(songsTwo, days)     

                (corr, pvalue) = st.pearsonr(ccOne, ccTwo)
                
                if figure:
                    rcParams['figure.figsize'] = 6, 4
                    fig = plt.figure()
                    ax = fig.add_subplot(111)
                    ax.plot(ccOne)
                    ax.plot(ccTwo)
                    string = 'corr:', corr, '- p-val:', pvalue
                    plt.title(string) 
                if corr > -2 and pvalue < pValue:
                    pvaluecount += 1
                    # nodig om 'inf' te ontwijken bij arctanh functie
                    if correctie:
                        if corr == -1:
                            corr = -0.9
                        elif corr == 1:
                            corr = 0.9
                    pearsonR.append(corr)
                break;
    if pCount:
        print(pvaluecount, songsOne['cc'], songsTwo['cc'])
    return pearsonR
```

```python
a = pearsonrCountry(usData, nlData, 1, False, True)
```

Waarbij het dicht bij de 0 zit als er totaal geen correlatie te vinden is en dichter bij de 1 of -1 komt als er een positieve of negatieve correlatie is. ## Correlatie Pearson-r grootste landen

Ik wil nu een overzichtelijk weergaven maken van de landen die allemaal onder elkaar worden correlateerd met minstens 30 nummer in de data. 

```python
query = db.npCountryTopCC.aggregate([
    {
        '$group': {
            '_id': {'cc': '$cc'},
            'count': {'$sum':1}
        }
    },
    {
        '$match': {'_id.cc': {'$ne': 'xx'}, 'count': {'$gt': 29} }
    },
    {
        '$project': {
            '_id': 0,
            'cc': '$_id.cc'
        }
    },
    {
        '$sort': {'cc': 1}
    }
])

topCountryCodes = list(query['result'])

countries = []
countryCodes = []
for countryCode in topCountryCodes:
    cc = countryCode['cc']
    countryCodes.append(cc)
    query = db.npCountryTopCC.find({'cc': cc})
    countryData = list(query)
    countries.append(countryData)
```

```python
def getPearsonPlotData(data, pValue, correctie, vCount = False):
    countData = []
    ccc = []
    for country in data:
        ccData = []
        for otherCountry in data:
            # zelfde land 
            if country[0]['cc'] != otherCountry[0]['cc']:
                b = arctanh(pearsonrCountry(country, otherCountry, pValue, correctie, False, vCount))
                scoreCC = tanh(b.mean())
                ccData.append(scoreCC)
            else:
                ccData.append(1)
        countData.append(ccData)
    return countData
```

```python
def displayPearsonPlot(data, countryCodes):
    rcParams['figure.figsize'] = 30, 20        
    pcolor(np.array(data),cmap='RdBu',vmin=1, vmax=-1)
    colorbar()

    yticks(arange(0.5,(len(countryCodes) +0.5)),countryCodes)
    xticks(arange(0.5,(len(countryCodes) +0.5)), countryCodes)
    show()
```

```python
d = getPearsonPlotData(countries, 1, False)
```

```python
displayPearsonPlot(d, countryCodes)
```

Hier de correlatie te zien van landen onder elkaar met minstens 30 muzieknummer in de top 50 muzieknummer. Wat opvalt is dat naast dat de bijvoorbeeld Australie naast zichzelf ook met bijvoorbeeld venusuela een perfect correlatie heeft. Dit is zeer onwaarschijnlijk en wil ik dus oplossen.
