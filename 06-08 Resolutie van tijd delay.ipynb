```python
%pylab inline

import numpy as np
import pymongo
import matplotlib.pyplot as plt
```

```python
days = [[104,111],[126,132],[137,139]] # DE UITIENDELIJK DAGEN DIE GEBRUIKT WORDEN!!!!!!

# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['mozart']

print('connected to database')
```

```python
query = db.npCountryTopCC.find({'title': 'Hey Mama'})

countries = list(query)
```

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

```python
def plotCountry(countryData, countriesCodes):
    for country in countryData:
        for cc in countriesCodes:
            if cc == country['cc'] or cc == 'ALL':
                plot(getCountryPlays(country, days), label=country['cc'])
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.title(country['title'])
```
