## Welke dagen ga ik pakken voor mij cross correlatie?

```python
def getArrayPlays(records, minDate, maxDate):
    daysIndex = np.arange(minDate, maxDate)
    arrayPlays = [];
    for dIndex in daysIndex:
        arrayPlays.append(records['value']['counts'][dIndex])
    return arrayPlays;
```

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

```python
days = [[97,111],[126,132],[137,139]]
for c in countries:
    plot(getCountryPlays(c, days))
```

```python
def plotSong(countryOne, countryTwo, mbId, norm):
    songsOne = None
    songsTwo = None
    days = [[104,111],[126,132],[137,139]]
    
    for cOne in countryOne:
        if cOne['mbId'] == mbId:
            songsOne = cOne
            break
    for cTwo in countryTwo:
        if cTwo['mbId'] == mbId:
            songsTwo = cTwo
            break
    
    ccOne = getCountryPlays(songsOne, days)
    ccTwo = getCountryPlays(songsTwo, days)
    if norm:
        ccOne = (ccOne - mean(ccOne)) / (std(ccOne) * len(ccOne))
        ccTwo = (ccTwo - mean(ccTwo)) / (std(ccTwo))
        
        cc = np.correlate(ccOne, ccTwo, "full")
        x=linspace(-round(len(cc)/2),round(len(cc)/2),len(cc))
        plt.plot(x,cc)
    else:
        plot(ccOne)
        plot(ccTwo)
```
