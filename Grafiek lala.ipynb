```python
%pylab inline

from pprint import pprint
import pymongo
import matplotlib.pyplot as plt
import time
import pandas as pd
```

```python
# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['mozart']

print('connected to database')
```

```python
tweetsRecords = db.testLocation.aggregate([
    {
        "$match": {'song.mb_id': {'$exists': 1} }
    },
    {
        "$project": {
            "_id": 0,
            "mb_id": '$song.mb_id',
            "created_at": '$tweet.created_at',
            "song": '$song.title'
        }
    },
    {
        "$group": {
           "_id": { "day": {"$dayOfMonth": "$created_at"}, "month": {"$month": "$created_at"}, "year": {"$year": "$created_at"}, "mb_id": "$mb_id"},
           "count": { "$sum": 1 },
           "song": {"$first": "$song"}
        }
    },
    {
        "$sort" : {'_id.mb_id': 1, "_id.year": 1, "_id.month": 1, "_id.day": 1}
    },
    { 
        "$project": {
            "_id": 0,
            "mb_id": "$_id.mb_id",
            "date": "$_id.day",
            "count": "$count",
            "song": "$song"
        }
    },
    { 
        "$group": {
            "_id": {"mb_id": "$mb_id"},
            "plays": { "$push": {"date": "$date", "count": "$count"}},
            "song": {"$first": "$song"}
        }
    },
    { 
        "$project": {
            "_id": 0,
            "plays": "$plays",
            "song": "$song",
            "playsCount": { "$size": "$plays" }
        }
    },
    {
        "$sort" : {'playsCount': -1}
    },
    { "$limit": 5 }
])

tweets = list(tweetsRecords['result'])
print("get records")
```
