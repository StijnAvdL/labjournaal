```python
def graphsCountrySongs(songsIndex, daysIndex, country, records):
    today = time.strftime("%d/%m/%Y")
    x=linspace(0,len(daysIndex),len(daysIndex))

    for sIndex in songsIndex:
        amount = [] 
        for dIndex in daysIndex:
            amount.append(records[sIndex]['value']['counts'][dIndex])
        line = plt.plot(x, amount, label=str((sIndex + 1)) + ': '+ records[sIndex]['title'])

    plt.ylabel('count plays')
    plt.locator_params(nbins=10)
    plt.xticks(x,daysIndex)
    plt.xlabel('dag in jaar 2015')
    plt.title('Song in ' + country + ': created at ' + today)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```
