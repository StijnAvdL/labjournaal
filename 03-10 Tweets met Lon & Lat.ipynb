Om geografische patronen te herkennen moet je weten waar de activiteit op de social media vandaan komt. Voor de Mozart project met de Top 40 maken we gebruik van de Twitter API om tweets binnen te halen die over de Top 40 muziek nummer gaan. Om de locatie te bepalen is de geo coördinatie het meest efficiënt. Daarom willen we weten hoeveel tweets dit veld ingevuld hebben.

```python
import pymongo

# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['app24236103']

totalTweets = db.top40plays.find().count()
print("Total tweets:", totalTweets)
```

```python
totalTweetsWithLoc = db.top40plays.find({'twitter_user_location.lat': {"$exists":"true"}}).count()
print(round((totalTweetsWithLoc/totalTweets)*100,2),"%")
```

Groot deel is van de tweets hebben een lon en lat. Dit is meer dan we verwacht hadden maar voor we met deze data gaan werken willen we eerst weten hoe erg betrouwbaar deze locatie zijn.

## Hoeveel % van de Tweets hebben een unieke lon en lat?

```python
totalUniqueTweetLoc = db.top40plays.aggregate([{'$group': { "_id": { 'lat': "$twitter_user_location.lat", 'lon': "$twitter_user_location.lon" } } }])

print(round((len(totalUniqueTweetLoc['result'])/totalTweetsWithLoc)*100,2),"%") 
```

Aanzienlijk deel is uniek maar nog steeds zijn we niet helemaal zeker of we deze data echt kunnen vertrouwen omdat we nog niet weten waar ze precies vandaan komen.
