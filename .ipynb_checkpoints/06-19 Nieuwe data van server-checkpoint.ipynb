Er is vanuit de server waar de #NP data wordt opgeslagen een aggregatie gedaan die de plays per country weergeven. Ik wil kijken of deze data betrouwbaar eruit ziet waarmee ik vervolgens aan de slag kan.

```python
%pylab inline

import numpy as np
import pymongo
import matplotlib.pyplot as plt
```

```python
# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['mozart']

print('connected to database')
```

```python
query = db.npCountry_aggr_3.aggregate([
    {
        '$match': {
            'title': {'$ne': None}
        }
    },
    {'$project': {
        'total': '$value.total',
        'mbId': '$mbId',
        'title': '$title'
        }
    },
    {
        '$group': {
            '_id': '$mbId',
            'total': {'$sum': '$total'},
            'title': {'$first': '$title'},
            'countCC': {'$sum': 1}
        }
    },
    {'$sort': {'total': -1}},
    {'$limit':500}
])

topNpSongs = list(query['result'])
print("Top 500 muzieknummers met meeste tweets")
```

```python
i = 0
for topNpSong in topNpSongs: 
    query = db.npCountry_aggr_3.find({'mbId': topNpSong['_id']})
    countries = list(query)
    
    for country in countries:
        db.npCountryTop500_3.insert({
            'cc': country['cc'],
            'mbId': country['mbId'],
            'value': country['value'],
            'title': country['title']
        })
print('done')
```

## Bekijken welke dagen nu data hebben verzameld

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

```python
days = [[90,170]]
```

```python
query = db.npCountryTop500_3.find({'title': 'Uptown Funk'})

countries = list(query)
for c in countries:
    plot(getCountryPlays(c, days))
```

```python
query = db.npCountryTop500_2.find({'title': 'Uptown Funk'})

countries = list(query)
for c in countries:
    plot(getCountryPlays(c, days))
```
