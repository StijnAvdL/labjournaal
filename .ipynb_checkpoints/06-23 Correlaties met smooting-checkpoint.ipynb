## Smooting data

```python
%pylab inline

import numpy as np
import pymongo
import matplotlib.pyplot as plt
import scipy.stats as st
import math
```

```python
days = [[126,132],[137,159]]  # DE UITIENDELIJK DAGEN DIE GEBRUIKT WORDEN!!!!!!

# Connection to Mongo DB
conn = pymongo.MongoClient('localhost', 27017)
db = conn['mozart']

print('connected to database')
```

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

```python
def getNoiseSignal(songData):
    # create linaire line with noise
    xdata=np.arange(0,len(songData), 1)
    increment, startValue = np.polyfit(xdata, songData, 1)
    lineary=startValue+increment*xdata
    
    #so what is the noise when we do a linear fit?
    yError=abs(songData-lineary)
    stdnoise=std(yError)    
    signaltonoiseratio=mean(songData)/stdnoise
    
    return signaltonoiseratio
```

```python
query = db.npCountryTop500_3.aggregate([
    { '$match': {'cc': 'US'} },
    { '$sort': {'value.total': -1}},
    {'$limit': 100}
])
usData = list(query['result'])
```

```python
def smooth(x, window_len=10, window='hanning'):
    if x.ndim != 1:
        raise(ValueError, "smooth only accepts 1 dimension arrays.")

    if x.size < window_len:
        raise(ValueError, "Input vector needs to be bigger than window size.")

    if window_len < 3:
        return x

    if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
        raise(ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'")

    s=np.r_[2*x[0]-x[window_len:1:-1], x, 2*x[-1]-x[-1:-window_len:-1]]
    #print(len(s))
    
    if window == 'flat': #moving average
        w = np.ones(window_len,'d')
    else:
        w = getattr(np, window)(window_len)
    y = np.convolve(w/w.sum(), s, mode='same')
    return y[window_len-1:-window_len+1]
```

```python
print(getNoiseSignal(getCountryPlays(usData[0], days)))
hanning = smooth(numpy.array(getCountryPlays(usData[0],days)),5,window="hanning")
print(getNoiseSignal(hanning))
print(getNoiseSignal(smooth(numpy.array(getCountryPlays(usData[0],days)),5,window="bartlett")))
print(getNoiseSignal(smooth(numpy.array(getCountryPlays(usData[0],days)),5,window="flat")))
print(getNoiseSignal(smooth(numpy.array(getCountryPlays(usData[0],days)),5,window="blackman")))
print(getNoiseSignal(smooth(numpy.array(getCountryPlays(usData[0],days)),5,window="hamming")))
```

```python
def plotSingalRuis(data):
    xdata=np.arange(0,len(data), 1)
    ydata = data
    regression = np.polyfit(xdata, ydata, 1)
    lineary=regression[1]+regression[0]*xdata
    return lineary
```

## Signaal-ruisverhouding

```python
def getNoiseCountry(countryCode, smoothLine=True):
    # get data
    query = db.npCountryTop500_3.aggregate([
        { '$match': {'cc': countryCode} },
        { '$sort': {'value.total': -1}}
    ])   
    countryData = list(query['result'])
    
    # edit data
    arrayNoise = []
    for data in countryData:
        if smoothLine:
            songData = smooth(numpy.array(getCountryPlays(data,days)),4,window="hamming")
        else:
            songData = getCountryPlays(data,days)
        arrayNoise.append(getNoiseSignal(songData))
    
    return arrayNoise
```

```python
usSingalRuisSmooth = getNoiseCountry('NL', True)
usSingalRuis = getNoiseCountry('NL', False)
```

```python
plot(usSingalRuisSmooth)
plot(usSingalRuis)
```

Je ziet duidelijk dat met de smooth de singnaal-ruisverhouding een stuk hoger komt te liggen

## Smooth met 'hamming'

```python
def getNoiseCountryData(countryCode):
    # get data
    query = db.npCountryTop500_3.aggregate([
        { '$match': {'cc': countryCode} },
        { '$sort': {'value.total': -1}}
    ])
    countryData = list(query['result'])
    
    # edit data
    arrayNoise = []
    for data in countryData:
        songData = getCountryPlays(data,days)
        songDataSmooth = smooth(numpy.array(getCountryPlays(data,days)),4,window="hamming")
        songDataSingalScore = getNoiseSignal(songDataSmooth)
        if songDataSingalScore > 4:
            fig = plt.figure()
            ax = fig.add_subplot(111)
            ax.plot(songData)
            ax.plot(songDataSmooth)
```
