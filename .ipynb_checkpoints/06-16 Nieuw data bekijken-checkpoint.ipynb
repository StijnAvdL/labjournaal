## Top 500 nieuwe data

Om niet telkens tussen alle gegevens te moeten kijken maak ik een aparte document met daarin de top 500 muzieknummers die in alle landen voorkomen.

```python
query = db.npCountry_aggr_2.aggregate([
    {
        '$match': {
            'title': {'$ne': None}
        }
    },
    {'$project': {
        'total': '$value.total',
        'mbId': '$mbId',
        'title': '$title'
        }
    },
    {
        '$group': {
            '_id': '$mbId',
            'total': {'$sum': '$total'},
            'title': {'$first': '$title'},
            'countCC': {'$sum': 1}
        }
    },
    {'$sort': {'total': -1}},
    {'$limit':500}
])

topNpSongs = list(query['result'])
print("Top 500 muzieknummers met meeste tweets")
```

```python
i = 0
for topNpSong in topNpSongs: 
    query = db.npCountry_aggr_2.find({'mbId': topNpSong['_id']})
    countries = list(query)
    
    for country in countries:
        db.npCountryTop500_2.insert({
            'cc': country['cc'],
            'mbId': country['mbId'],
            'value': country['value'],
            'title': country['title']
        })
print('done')
```

## Welke dagen is er data verzameld

Nu er een kleinere selectie is gemaakt van de data ben ik benieuwd welke dagen precies nu te gebruiken zijn voor correlatie functies en signaal-ruisverhouding. Hierbij pak ik niet één uniek land maar één uniek muzieknummer omdat ik zeker wil weten dat in alle landen de ontwikkelingen in verhouding gelijk lopen.

```python
def getArrayPlays(records, minDate, maxDate):
    daysIndex = np.arange(minDate, maxDate)
    arrayPlays = [];
    for dIndex in daysIndex:
        arrayPlays.append(records['value']['counts'][dIndex])
    return arrayPlays;
```

```python
def getCountryPlays(data, days):
    arrayPlays = [];
    for day in days:
        dayIndexes = np.arange(day[0], day[1])
        for dayIndex in dayIndexes:
            arrayPlays.append(data['value']['counts'][dayIndex])
    return arrayPlays
```

Landen induvueel laten op na Nederland zien dat op alle dagen ze ongeveer even vaak gevonden zijn. 

```python
query = db.npCountryCount.find()
countData = list(query)
i = 0
countDataArray = []
while i <= 365:
    totalCount = 0
    for cData in countData:
        totalCount += cData['value']['counts'][i]
    countDataArray.append(totalCount)
    i += 1  
```

```python
days = [[90,165]]
arrayPlays = [];
for day in days:
    dayIndexes = np.arange(day[0], day[1])
    for dayIndex in dayIndexes:
        arrayPlays.append(countDataArray[dayIndex])
plot(arrayPlays)
```

Het aantal gevonden velden zijn over het algemeen gelijk gebleven bij de laatste 2 pieken maar de eerste is duidelijk een stuk lager.

```python
query = db.npCountryCount.find({'_id.cc': 'FR'})
data = list(query)
totalCount = 0
for d in data[0]['value']['counts']:
    totalCount += d
print(totalCount)
```
